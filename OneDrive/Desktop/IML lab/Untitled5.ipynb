```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
x = [2,4,6,8,10]
y = [3,6,8,11,14]

data = {'X':x , 'Y':y}
df = pd.DataFrame(data)
print(df)
```

```python
%matplotlib inline
plt.scatter(df.X,df.Y)
plt.xlabel('C')
plt.ylabel('Units')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['X']],df.Y)
```
