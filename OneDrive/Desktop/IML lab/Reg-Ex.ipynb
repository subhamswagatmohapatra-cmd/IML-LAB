```python
import re
pattern = "good"
text = "deba is a good guy, deba is very humble"

match = re.search(pattern,text)
print(match)
```

```python
import re
string = "hello 12 hi 89. Howdy 34"
pattern = '\d-'

result = re.findall(pattern,string)
print(result)
```

```python
import re
pattern = r"dob2"
text = "DOB1: 12-06-1997, DOB2: 20-05-2002"

match = re.search(pattern,text,re.IGNORECASE)
print(match)
```
