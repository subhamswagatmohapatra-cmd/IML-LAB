```python
import numpy as np
```

```python
x1 = np.array([1,2,3])
y1 = np.array([2,4,5])
y2 = np.array([10,20,25])
```

```python
x = np.column_stack((np.ones(3),x1))
```

```python
y = np.column_stack((y1,y2))
```

```python
B = np.linalg.inv(x.T@x) @x.T@y
```

```python
b00 = B[0,0]
b01 = B[0,1]
b10 = B[1,0]
b11 = B[1,1]
```

```python
print(f"b00 = {b00}")
print(f"b01 = {b01}")
print(f"b10 = {b10}")
print(f"b11 = {b11}")
```
