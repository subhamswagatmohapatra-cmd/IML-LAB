```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


data = np.array([
    [2,10],
    [2,5],
    [8,4],
    [5,8],
    [7,5],
    [6,4],
    [1,2],
    [4,9]
])

point_names = ['a1','a2','a3','b1','b2','b3','c1','c2']

k = 3
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_


print("---K-Means Numerical Results---")
print(f"{'Point':<6} | {'Coordinates':<10} | {'Cluster Assignment'}")
print("-" * 40)
for i in range(len(data)):
  print(f"{point_names[i]:<6} | {str(data[i]):<10} | Cluster {labels[i]}")

print("\n---Final Centroids---")
for i,center in enumerate(centroids):
  print(f"Cluster {i} Centroid: {center}")


plt.scatter(data[:,0],data[:,1],c = labels,cmap = 'viridis',s = 50)
plt.scatter(centroids[:,0],centroids[:,1],c = 'black',s = 200,marker = 'x')
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
