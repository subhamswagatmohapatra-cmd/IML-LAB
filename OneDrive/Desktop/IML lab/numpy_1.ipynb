```python
import numpy as np
```

```python
x1 = np.array([0,1,2,3,])
x2 = np.array([0,1,4,9])
y = np.array([1.1,1.9,5.2,9.8])
```

```python
x = np.column_stack((np.ones(4),x1,x2))
```

```python
B = np.linalg.inv(x.T@x) @x.T@y
```

```python
b0,b1,b2 = B
```

```python
print("Intercept (b0)",b0)
print("Coefficient (b1)",b1)
print("Coefficient (b2)",b2 )
```
