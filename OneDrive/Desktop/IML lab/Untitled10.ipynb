```python
import numpy as np
```

```python
x1 = np.array([1,2,3])
y = np.array([2,4,5])
```

```python
x = np.column_stack((np.ones(3),x1))
```

```python
B = np.linalg.inv(x.T@x) @x.T@y
```

```python
b0,b1, = B
```

```python
print("Intercept(b0)",b0)
print("Coefficient for Studied Hours(b1)",b1)
```
