```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
x = [1,2,3,4,5]
y = [10,9,7,6,4]

data = {'X':x , 'Y':y}
df = pd.DataFrame(data)
print(df)
```

```python
%matplotlib inline
plt.scatter(df.X,df.Y)
plt.xlabel('KG')
plt.ylabel('Defects')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['X']],df.Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.scatter(df.X,df.Y)
plt.xlabel('KG')
plt.ylabel('Defects')
plt.plot(df.X,reg.predict(df[['X']]),color='blue')
```
