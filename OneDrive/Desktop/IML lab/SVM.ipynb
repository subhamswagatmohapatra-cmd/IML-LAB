```python
import numpy as np
from sklearn import svm

x = np.array([
    [2,2],
    [4,5],
    [7,4]
])

y = np.array([-1,1,1])

clf = svm.SVC(kernel='linear',C=100)
clf.fit(x,y)

weights = clf.coef_[0]
intercept = clf.intercept_[0]
support_vectors = clf.support_vectors_

print("-- SVM Numerical---")
print(f"Weights(w): {weights}")
print(f"Intercept(b): {intercept}")
print(f"Support Vectors: {support_vectors}")

print(f"\n Decision Boundary Equation: {weights[0]:.2f}x1 + {weights[1]:.2f}x2 + {intercept}")
```
