```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
x = [1,2,3,4,5]
y = [2,3,5,5,8]

data = {'X':x , 'Y':y}
df = pd.DataFrame(data)
print(df)
```

```python
%matplotlib inline
plt.scatter(df.X,df.Y)
plt.xlabel('Ad Spends')
plt.ylabel('Sales')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['X']],df.Y)
```
