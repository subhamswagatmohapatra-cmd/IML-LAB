```python
import math
```

```python
point=['a','b','c','d']
x1=[7,8,5,9]
x2=[2,1,4,6]
p=[8,2]
k=3
distances=[]
for i in range(len(point)):
  dist=math.sqrt((p[0]-x1[i])**2+(p[1]-x2[i])**2)
  print(point[i],dist)
  distances.append((point[i],dist))
```

```python
distances.sort(key=lambda x:x[1])
```

```python
neighbors=distances[:k]
```

```python
print(f"The {k} nearest neighbors to point {p} are:")
for name,distances in neighbors:
  print(f"point {name} with distance{dist:.2f}")
```

```python
import math
from collections import Counter
```

```python
points=['A','B','C','D']
x1=[7,8,5,9]
x2=[2,1,4,6]
labels=['pos','neg','pos','neg']
p=[8,2]
k=3
```

```python
dataset=[]
for i in range(len(points)):
  dist=math.sqrt((p[0]-x1[i])**2 +(p[1]-x2[i])**2)
  dataset.append({'name':points[i],'dist':dist,'label':labels[i]})
```

```python
dataset.sort(key=lambda x:x['dist'])
```

```python
neighbors=dataset[:k]
```

```python
neighbors_labels=[n['label'] for n in neighbors]
prediction=Counter(neighbors_labels).most_common(1)[0][0]
```

```python
print(f"Target Point p{p} Results:")
print("-" *30)
for n in neighbors:
    print(f"Neighbors{n['name']}:Distance{n['dist']:.2f},Class:{n['label']}")
print("-" *30)
print(f"final Prediction for p:{prediction}")
```
